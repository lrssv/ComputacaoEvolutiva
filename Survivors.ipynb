# Model for classifying the Perpetrator's gender using an evolutionary algorithm to select features.

## - Survivors

This code has all steps for Survivors. Where:

* Imports: the necessary libraries, parents and children;
* Scripts: separes 5 best parents and 95 best children for next generation;
* Setups: results.

## Imports

```
import pickle
import pandas as pd
import numpy as np
import random
```

```
with open('/content/drive/My Drive/Data Files/fitness.txt', "rb") as fp:   
  population = pickle.load(fp)
```

```
with open('/content/drive/My Drive/Data Files/mutandcross.txt', "rb") as fp:   
  generation = pickle.load(fp)
```

## Scripts

```
def survivors(generation,population):
  count1,count2 = 0,0
  surv = []

  while count1 < 95:
    child1 = random.choice(generation) 
    child2 = random.choice(generation)

    if child1[1] > child2[1]:
      surv.append(child1)
    else:
      surv.append(child2)  
    
    count1 += 1

  while count2 < 2:
    parent1 = random.choice(population) 
    parent2 = random.choice(population)
      
    if parent1[1] > parent2[1]:
      surv.append(parent1)
    else:
      surv.append(parent2)
    count2 += 1
        
  return surv
```

## Setups

```
survs = survivors(generation,population)
```
